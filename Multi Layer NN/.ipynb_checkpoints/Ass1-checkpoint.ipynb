## Evaluation

```python
# def softmax(x):
#     #Instead of using np.exp(x)/np.sum(np.exp(x))
#     #I decrease the value of x with the max, 
#     #it could be any number as it cancels out when equation is expanded
#     #This is to avoid overflow due to exponential increase
#     e_x = np.exp(x - np.max(x))
#     soft= e_x / np.sum(e_x,axis=0)
#     return soft

def softmax( x):
    try:
        e = np.exp(x)
        return e / np.sum(e, axis=0)
    except FloatingPointError:
        # What happens here is that we get some of the values in x that are too big
        # so the exp overflows. In that case we check column by column where the
        # problem is
        return softmax_column_by_column(x)

def softmax_column_by_column( x):
    cols = x.shape[1]
    res = np.finfo(float).eps * np.ones_like(x)
    for c in range(cols):
        try:
            e = np.exp(x[:, c])
            res[:, c] = e / np.sum(e, axis=0)
        except FloatingPointError:
            res[np.argmax(x[:, c]), :] = 1 - np.finfo(float).eps
    return res

def EvaluateClassifier(X, W, b):
    """
    s = W x + b
    p = SOFTMAX (s) 
    """
    s = np.dot(W,X) + b
    P = softmax(s)
    #Dimensional check
    assert(X.shape == (d,N))
    assert(P.shape == (K,N))
    return P
```

#### Section 4

## Computing the cost

```python
def ComputeCost_CrossEntropy(P,Y,W):
 
    P = np.reshape(P, (K, -1))
    ground_truth = np.reshape(Y, (K, -1))
    product = np.multiply(Y, P).sum(axis=0)
    
    #cross entropy loss - Handling -log0 tending to infinity
    product[product == 0] = np.finfo(float).eps    #very low value
    crossEntropyLoss= - np.log(product).sum() / N
    L2_regularization_cost=lambd * np.power(W, 2).sum()
    
    return  crossEntropyLoss + L2_regularization_cost
```

```python
cost=ComputeCost_CrossEntropy(P,Y,W)
print("cost :"+str(cost))
```

#### Section 5

## Accuracy of the predictions

```python
def ComputeAccuracy(P,Y):
    #instead of using function acc = ComputeAccuracy(X, y, W, b)
    # and calculating P again lets use it from the from the prev. output
    predictions=np.argmax(P,axis=0)
    groundtruth=np.argmax(Y,axis=0)
    matches=np.sum(predictions==groundtruth)
    total=len(predictions)
    accuracy=(matches/float(total))
    return accuracy
```

```python
accuracy=ComputeAccuracy(P,Y)
print ("Accuracy :"+ str(accuracy))
```

#### Section 6

## Computing the Gradients

```python
def ComputeGradients(X, Y, P, W, lambd):
    """ 
    G -- dJ/dZ - g for Batch
    Z -- WX
    grad_W1 -- dJ/dW
    grad_b -- dJ/db
    grad_L -- dJ/dL Regularization term
    grad_W -- dJ/dW + Regularization term

    lambd - lambda, to differ from the keyword itself
    """
    grad_w = np.zeros_like(W)
    grad_b = np.zeros_like(b)
    N_float=float(X.shape[1])
    N=X.shape[1]
    
    G= P-Y
    grad_W1= (1/N)*np.dot(G,X.T)
    grad_L = 2*lambd*W
    grad_W = grad_W1+grad_L
    grad_b = (1/N)*np.dot(G,np.ones((int(N),1)))
    grad_b = np.sum(G, axis=1, keepdims=True)/ N
    
    for i in range(N):
            x = X[:, i]
            y = Y[:, i]
            p = P[:, i]

            g = p - y

            grad_w += np.outer(g, x)
            grad_b += np.reshape(g, grad_b.shape)

    grad_w = grad_w / N + 2 * lambd * W
    grad_b /= N
    
    assert(grad_W.shape == (K,d))
    assert(grad_b.shape == (K,1))
    
    return grad_W, grad_b
```

```python
grad_W,grad_b=ComputeGradients(X,Y,P,W,lambd)
plt.rcParams['figure.figsize'] = [7, 2]
plt.subplot(121).set_title("W")
ax = sns.heatmap(grad_W)
plt.subplot(122).set_title("b")
ax=sns.heatmap(grad_b)
```

```python
def compute_grads_num(X, Y,W,b):
    h = np.finfo(float).eps

    grad_W = np.zeros_like(W)
    grad_b = np.zeros_like(b)

    P = EvaluateClassifier(X,W,b)
    c1 = ComputeCost_CrossEntropy(P, Y,W)

    for i in range(b.shape[0]):
        b[i] += h
        P = EvaluateClassifier(X,W,b)
        c2 = ComputeCost_CrossEntropy(P, Y,W)
        grad_b[i] = (c2 - c1) / h
        b[i] -= h

    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            W[i, j] += h
            P = EvaluateClassifier(X,W,b)
            c2 = ComputeCost_CrossEntropy(P, Y,W)
            grad_W[i, j] = (c2 - c1) / h
            W[i, j] -= h

    return grad_W, grad_b

def gradient_check(X,Y,W,b, epsilon=1e-6):

    gradWnum,gradbnum=compute_grads_num(X,Y,W,b)
    gradW,gradb =ComputeGradients(X,Y,P,W,lambd)
    
    grad=gradW
    gradapprox=gradWnum
    
    numerator = np.linalg.norm(grad - gradapprox)     
    denominator = np.linalg.norm(grad) + np.linalg.norm(gradapprox)    
    difference = numerator / denominator                               
    
    if difference < 1e-6:
        print("The gradient is correct!")
    else:
        print("The gradient is wrong!")
        print("Variation of :"+str(difference))
    
    print_grad_diff(gradW, gradWnum, gradb, gradbnum) 
    
    return difference

def print_grad_diff(grad_w, grad_w_num, grad_b, grad_b_num):
    print('Grad W:')
    print('- sum of abs differences: {:.3e}'.format(np.abs(grad_w - grad_w_num).sum()))
    print('- mean of abs values   W: {:.3e}   W_num: {:.3e}'
          .format(np.abs(grad_w).mean(), np.abs(grad_w_num).mean()))
    print('- min  of abs values   W: {:.3e}   W_num: {:.3e}'
          .format(np.abs(grad_w).min(),  np.abs(grad_w_num).min()))
    print('- max  of abs values   W: {:.3e}   W_num: {:.3e}'
          .format(np.abs(grad_w).max(),  np.abs(grad_w_num).max()))
    print('Grad b:')
    print('- sum of abs differences: {:.3e}'.format(np.abs(grad_b - grad_b_num).sum()))
    print('- mean of abs values   b: {:.3e}   b_num: {:.3e}'
          .format(np.abs(grad_b).mean(), np.abs(grad_b_num).mean()))
    print('- min  of abs values   b: {:.3e}   b_num: {:.3e}'
          .format(np.abs(grad_b).min(),  np.abs(grad_b_num).min()))
    print('- max  of abs values   b: {:.3e}   b_num: {:.3e}'
          .format(np.abs(grad_b).max(),  np.abs(grad_b_num).max()))
```

```python
def update_params(W,b,grad_W,grad_b,eta):
    W=W-eta*grad_W
    b=b-eta*grad_b
    return W,b
```

```python
#Training
for i in range(10000):
        P=EvaluateClassifier(X,W,b)
        # Get the cost of the predictions
        cost=ComputeCost_CrossEntropy(P, Y, W)
        # Get accuracy of predictions
        acc=ComputeAccuracy(P,Y)
        # Learn the gradients of the cost function
        grad_W,grad_b=ComputeGradients(X,Y,P,W,lambd)
        # update params
        W,b=update_params(W,b,grad_W,grad_b,0.001)
        print('Step '+str(i)+": Cost:"+str(cost)+" Accuracy: "+str(acc)) 
```

```python
def get_mini_batches(X, Y, n_batch): 
    mini_batches = []
    X=X.T
    Y=Y.T
    data = np.hstack((X, Y)) 
    np.random.shuffle(data) 
    n_minibatches = data.shape[0] // n_batch 
    i = 0

    for i in range(n_minibatches - 1): 
        mini_batch = data[i * n_batch:(i + 1)*n_batch, :] 
        X_mini = mini_batch[:, :-10] 
        Y_mini = mini_batch[:, 3072:].reshape((100,10))
        mini_batches.append((X_mini, Y_mini)) 

    if data.shape[0] % n_batch != 0: 
        mini_batch = data[i * n_batch:data.shape[0]] 
        X_mini = mini_batch[:, :-10] 
        Y_mini = mini_batch[:, 3072:].reshape((100,10))
        mini_batches.append((X_mini, Y_mini)) 

    return mini_batches 

def MiniBatchGD(X,Y,GDparams,W,b,lambd):
    step=0
    for i in range(GDparams['n_epochs']):
        print (i)
        mini_batches=get_mini_batches(X, Y, GDparams['n_batch'])
        #print ("Processing mini batches " + u'\u2713')
        for batch in mini_batches:
            #do transpose again 
            X_b=batch[0].T
            Y_b=batch[1].T
            
            # Get some predictions
            P=EvaluateClassifier(X_b,W,b)
            # Get the cost of the predictions
            cost=ComputeCost_CrossEntropy(P, Y_b, W)
            # Get accuracy of predictions
            acc=ComputeAccuracy(P,Y)
            # Learn the gradients of the cost function
            grad_W,grad_b=ComputeGradients(X_b,Y_b,P,W,lambd)
            # update params
            W,b=update_params(W,b,grad_W,grad_b,GDparams['eta'])
            print('Step '+str(step)+": Cost:"+str(cost)+" Accuracy: "+str(acc))
            step=step+1
        #print("Epoch "+str(i+1)+ u'\u2713')
```
